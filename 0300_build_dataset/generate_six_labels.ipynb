```python
import pandas as pd
from tensorflow.python.lib.io.file_io import FileIO as open_file
import random
import csv
```

```python
with open_file("gs://alex-s2t-test/adversarial/datasets/original/data_set.csv", "r") as f:
    df= pd.read_csv(f)
df
```

```python
df["labels"].value_counts()
```

```python
max_images = df["labels"].value_counts()["AAAPistol"]
max_images
```

```python
pistol=df[df["labels"]=="AAAPistol"]["image"].unique()
knives=random.sample(df[df["labels"]=="KNIVES"]["image"].unique(),max_images)
airplane=random.sample(df[df["labels"]=="airplanes"]["image"].unique(),max_images)
motorbike=random.sample(df[df["labels"]=="Motorbikes"]["image"].unique(),max_images)
face=random.sample(
    list(df[df["labels"]=="Faces_easy"]["image"].unique())+list(df[df["labels"]=="Faces"]["image"].unique()),
    max_images
)
```

```python
other_labels=[x for x in df["labels"].unique() if x not in 
              ["AAAPistol","KNIVES","airplanes","Motorbikes","Faces_easy","Faces"]
             ]
other_labels
```

```python
other = random.sample(df[df["labels"].isin(other_labels)]["image"].unique(),max_images)
```

```python
len(pistol),len(knives),len(airplane),len(motorbike),len(face),len(other)
```

```python
out_content=[
    [x,"pistol"] for x in pistol
]+[
    [x,"knife"] for x in knives 
]+[
    [x,"airplane"] for x in airplane 
]+[
    [x,"motorbike"] for x in motorbike 
]+[
    [x,"face"] for x in face 
]+[
    [x,"other"] for x in other 
]
out_df=pd.DataFrame(out_content,columns=["image","label"])
```

```python
out_df
```

```python
with open_file("gs://alex-s2t-test/adversarial/datasets/six_labels/labels/data_set.csv", "w") as f:
    out_df.to_csv(f, index=False, quoting=csv.QUOTE_ALL)
```

```python
num_eval=int(max_images*0.1)
num_train = max_images - num_eval
```

```python
train_content=[
    [x,"pistol"] for x in pistol
][:num_train]+[
    [x,"knife"] for x in knives 
][:num_train]+[
    [x,"airplane"] for x in airplane 
][:num_train]+[
    [x,"motorbike"] for x in motorbike 
][:num_train][:num_train]+[
    [x,"face"] for x in face 
][:num_train]+[
    [x,"other"] for x in other 
]
train_df=pd.DataFrame(train_content,columns=["image","label"])
eval_content=[
    [x,"pistol"] for x in pistol
][num_train:]+[
    [x,"knife"] for x in knives 
][num_train:]+[
    [x,"airplane"] for x in airplane 
][num_train:]+[
    [x,"motorbike"] for x in motorbike 
][num_train:]+[
    [x,"face"] for x in face 
][num_train:]+[
    [x,"other"] for x in other 
]
eval_df=pd.DataFrame(eval_content,columns=["image","label"])
```

```python
with open_file("gs://alex-s2t-test/adversarial/datasets/six_labels/labels/train_set.csv", "w") as f:
    train_df.sample(frac=1).to_csv(f, index=False, quoting=csv.QUOTE_ALL)
with open_file("gs://alex-s2t-test/adversarial/datasets/six_labels/labels/eval_set.csv", "w") as f:
    eval_df.sample(frac=1).to_csv(f, index=False, quoting=csv.QUOTE_ALL)
```

```python
with open_file("gs://alex-s2t-test/adversarial/datasets/six_labels/labels/labels.txt", "w") as f:
    f.write("\n".join(sorted(out_df["label"].unique())))
```
