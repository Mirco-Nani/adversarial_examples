```python
import tensorflow as tf
from google.protobuf.json_format import MessageToJson
import json


results = []


for i in range(7): 
    tfrecord_file="gs://alex-s2t-test/adversarial/datasets/six_labels/inception_resnet/preproc/tfrecord/train-0000"+str(i)+"-of-00007.tfrecord.gz"
    for example in tf.python_io.tf_record_iterator(tfrecord_file, tf.python_io.TFRecordOptions(tf.python_io.TFRecordCompressionType.GZIP)):
        results.append( json.loads( MessageToJson(tf.train.Example.FromString(example)) ) )

    
len(results)
```

```python
labels=[]
for x in results:
    if "label" in x["features"]["feature"]:
        labels.append(x["features"]["feature"]["label"]["int64List"]["value"][0])
```
